```python
# load the pandas library
import pandas as pd


filename = "https://raw.githubusercontent.com/YALINYAN-YU/J-J-data/main/patients.csv"
df3 = pd.read_csv(filename)
df3.head()
```

```python
import numpy as np, pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df3.describe()
```

```python
# See the distribution of gender to recognize different distributions
sns.countplot(x='gender', data=df3)
plt.title('Distribution of Gender')
```

```python
# Histogram for age distribution
df3.hist('age', bins=45, alpha=0.8)
plt.title('Distribution of Age')
plt.xlabel('Age')
```

```python
# Histogram for age distribution
df3.hist('age', bins=5, alpha=0.8)
plt.title('Distribution of Age')
plt.xlabel('Age')
```

```python
#Histogram for age distribution by gender
plt.hist('age', data=df3[df3['gender'] == 'Male'], alpha=0.5, label='Male');
plt.hist('age', data=df3[df3['gender'] == 'Female'], alpha=0.5, label='Female');
plt.title('Distribution of Age by Gender');
plt.xlabel('Age');
plt.legend();
```
